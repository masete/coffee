### Top Producers: Identify the top coffee-producing countries for each year. This could provide insights into shifts in global coffee production leadership

```python
# Raw data as provided by the user
data = """
Country\t2018/2019\t2019/2020\t2020/2021\t2021/2022\t2022/2023\t2023/2024
Brazil\t4,023\t4,040\t3,954\t4,055\t3,600\t4,300
Vietnam\t2,150\t2,250\t2,300\t2,350\t2,350\t2,500
Malaysia\t3,125\t3,000\t2,780\t2,975\t2,800\t2,300
European Union\t1,160\t1,215\t1,460\t1,820\t2,000\t2,200
India\t1,838\t1,782\t1,970\t2,310\t2,050\t2,080
Thailand\t730\t905\t885\t990\t1,000\t1,100
Indonesia\t1,200\t1,000\t1,350\t1,050\t1,040\t1,000
Colombia\t900\t1,025\t1,010\t1,100\t900\t1,000
Mexico\t943\t945\t865\t695\t880\t900
Ecuador\t410\t370\t475\t413\t465\t400
China\t350\t230\t300\t315\t325\t350
Cote d'Ivoire\t240\t290\t285\t325\t300\t300
Nicaragua\t110\t120\t75\t75\t75\t75
Guatemala\t4\t5\t11\t15\t15\t15
El Salvador\t14\t12\t15\t11\t10\t10
Tanzania\t30\t30\t7\t7\t10\t10
"""

# # df = pd.read_csv(pd.compat.StringIO(data), delimiter='\t')

# # Convert the data to a DataFrame
# df = pd.read_csv(pd.compat.StringIO(data), delimiter='\t')

# # Set the "Country" column as the index
# df.set_index("Country", inplace=True)

# # Remove commas and convert to numeric
# df = df.apply(lambda x: x.str.replace(',', '').str.replace('\t', '')).astype(float)

# # Calculate z-scores
# z_scores = (df - df.mean()) / df.std()

# # Plotting the data
# plt.figure(figsize=(12, 6))

# for country in z_scores.index:
#     plt.plot(z_scores.columns, z_scores.loc[country], marker='o', label=country)

# plt.axhline(y=2, color='r', linestyle='--', label='2 SD')
# plt.axhline(y=-2, color='r', linestyle='--')

# plt.xlabel("Year")
# plt.ylabel("Z-Score")
# plt.title("Z-Scores of Coffee Production Anomalies")
# plt.legend()
# plt.grid(True)
# plt.xticks(rotation=45)
# plt.tight_layout()

# plt.show()

# Convert the data to a DataFrame
lines = data.strip().split('\n')
header = lines[0].split('\t')
lines = lines[1:]
data_list = [line.split('\t') for line in lines]
df = pd.DataFrame(data_list, columns=header)
df.set_index("Country", inplace=True)

# Remove commas and convert to numeric
df = df.apply(lambda x: x.str.replace(',', '').str.replace('\t', '')).astype(float)

# Calculate z-scores
z_scores = (df - df.mean()) / df.std()

# Plotting the data
plt.figure(figsize=(12, 6))

for country in z_scores.index:
    plt.plot(z_scores.columns, z_scores.loc[country], marker='o', label=country)

plt.axhline(y=2, color='r', linestyle='--', label='2 SD')
plt.axhline(y=-2, color='r', linestyle='--')

plt.xlabel("Year")
plt.ylabel("Z-Score")
plt.title("Z-Scores of Coffee Production Anomalies")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
