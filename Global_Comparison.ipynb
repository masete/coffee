```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('global comparison of instant coffee sales.xlsx')
```

```python
df1.head()
```

```python
# Assuming df is your DataFrame
df1 = df1.iloc[2:].reset_index(drop=True)
```

```python
df1.head()
```

```python
# Save the first row as the new column names
new_columns = df1.iloc[0]
```

```python
# Check if there's a NaN value in new_columns, and replace it with 'country'
if pd.isna(new_columns).any():
    new_columns = new_columns.fillna('country')
```

```python
# Drop the first row (index 0) and all columns
df1 = df1.drop(df1.index[0], axis=0).reset_index(drop=True)
```

```python
# Set the new column names
df1.columns = new_columns
```

```python
df1.head()
```

### Line chart showing how Sales have evolved from 2012-2025

#### Japan

```python
# Assuming df is your original DataFrame
# Filter rows where the 'country' column is 'Japan'
japan_data = df1[df1['country'] == 'Japan']

# Drop the 'Rank' and 'country' columns, leaving only the years and sales data
japan_mini_df = japan_data.drop(columns=['Rank', 'country'])

# Print the mini DataFrame for Japan
print(japan_mini_df)
```

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(japan_mini_df.columns.astype(str), japan_mini_df.values[0], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Japan - Instant Coffee Sales Trend')
plt.grid(True)
plt.xticks(japan_mini_df.columns.astype(str))
plt.show()
```

```python
# Convert columns to numeric values
df = japan_mini_df.apply(pd.to_numeric, errors='coerce')

# Function to calculate growth rate for each year
def calculate_growth_rate(row):
    growth_rate = row.diff() / row.shift()
    return growth_rate

# Calculate growth rate for each year
growth_rate_df = df.iloc[:, 3:].apply(calculate_growth_rate, axis=1)

# Calculate average growth rate
average_growth_rate = growth_rate_df.mean(axis=1)[0]

# Calculate total growth rate
total_growth_rate = growth_rate_df.sum(axis=1)[0]

# Determine the year with the highest growth rate
highest_growth_year = growth_rate_df.iloc[0].idxmax()  # Get the column name with the highest growth rate

print("Average Growth Rate:", average_growth_rate)
print("Total Growth Rate:", total_growth_rate)
print("Highest Growth Rate Year:", highest_growth_year)
```

### Grouped data

```python
# List of chosen countries
chosen_countries = ['Uganda', 'Ethiopia', 'Sudan', 'South Korea', 'Japan', 'United States', 'China', 'Brazil', 'Singapore', 'Netherlands']

# Convert the year to a string (e.g., '2012')
year_to_plot = '2012'
data_year = df1[['country', year_to_plot]]

# Filter the data for the chosen countries
data_year = data_year[data_year['country'].isin(chosen_countries)]

# Convert the sales data to numeric values
data_year[year_to_plot] = pd.to_numeric(data_year[year_to_plot], errors='coerce')

# Drop rows with NaN values in the sales data
data_year.dropna(subset=[year_to_plot], inplace=True)

# Sort the data by sales in descending order
data_year = data_year.sort_values(by=year_to_plot, ascending=False)
```

```python
# Plot the grouped bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_year['country'], data_year[year_to_plot])
plt.xlabel('Country')
plt.ylabel('Sales')
plt.title(f'Instant Coffee Sales in {year_to_plot} - Chosen Countries')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### Growth Rates

```python
# Filter the data for the countries of interest
countries_of_interest = ['Uganda', 'Ethiopia', 'Sudan', 'South Korea', 'Japan', 'United States', 'China', 'Brazil', 'Singapore', 'Netherlands']
data_countries_of_interest = df1[df1['country'].isin(countries_of_interest)]

# Transpose the DataFrame to have years as rows and countries as columns
data_countries_of_interest = data_countries_of_interest.set_index('country').T

# Convert the sales data to numeric values
data_countries_of_interest = data_countries_of_interest.apply(pd.to_numeric, errors='coerce')

# Calculate the growth rate for each country from year to year
growth_rate = data_countries_of_interest.pct_change() * 100

# Plot the growth rate for each country
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    plt.plot(growth_rate.index.astype(str), growth_rate[country], marker='o', label=country)

plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.title('Growth Rate of Instant Coffee Sales by Country')
plt.legend()
plt.grid(True)
plt.show()

# Calculate average growth rate, total growth rate, highest growth year, and lowest growth year for each country
average_growth_rate = growth_rate.mean()
total_growth_rate = data_countries_of_interest.iloc[-1] / data_countries_of_interest.iloc[0] * 100 - 100
highest_growth_year = growth_rate.idxmax()

# Display the calculations for each country
for country in countries_of_interest:
    print(f"\nMetrics for {country}:")
    print("Average Growth Rate:", average_growth_rate[country])
    print("Total Growth Rate:", total_growth_rate[country])
    # print("Year with Highest Growth Rate:", highest_growth_year[country])
```
